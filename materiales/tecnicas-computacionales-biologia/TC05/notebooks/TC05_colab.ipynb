## 1. Descargar y verificar el paquete

El paquete `TC05_starter_v0.1.0.tar.gz` vive en el repositorio público `Asn4code/pid_2026`. Se descarga, se comprueba su `SHA256SUMS` y se extrae.

```python
from pathlib import Path
from urllib.request import urlopen

BASE_URL = "https://github.com/Asn4code/pid_2026/raw/desarrollo/materiales/tecnicas-computacionales-biologia/TC05/paquete"
for nombre in ("TC05_starter_v0.1.0.tar.gz", "SHA256SUMS"):
    destino = Path("/content") / nombre
    with urlopen(f"{BASE_URL}/{nombre}") as respuesta:
        destino.write_bytes(respuesta.read())
    print("descargado", nombre)
```

```python
%%bash
set -euo pipefail
cd /content
sha256sum -c SHA256SUMS
tar -xzf TC05_starter_v0.1.0.tar.gz
ls TC05-starter
```
